## load neccessary libaries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# basic graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
## resize your graph
plt.figure(figsize=(6,4),dpi=700)


## line number two

x2=np.arange(0,4.5,0.5)
           
           
plt.plot(x2,x2**2,'r',label='x^2')
plt.plot(x,y,label='2x',color='grey',linewidth=2,marker='*',markersize=10)        
plt.title('our first graph',fontdict={'fontname': 'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')




plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.savefig('mygraph.png',dpi=700)

plt.legend()

plt.show()
```

###barchart

```python
values=[8,4,7]
labels=['A','B','C']

bars=plt.bar(labels,values)

bars[0].set_hatch('*')
bars[1].set_hatch('+')
bars[2].set_hatch('*')

plt.figure(figsize=(3,7),dpi=400)
plt.show()

```

##line graph

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

```python
plt.plot(gas.Year,gas.USA)
plt.show()
```

```python
## you can make the graph look bigger
plt.figure(figsize=(6,4),dpi=300)

## you can add titles by
plt.title('gas prices in first word contries(in usd)')
## you can add more than one country to the graph and give them labels
plt.plot(gas.Year,gas.Germany,label='Germany',color='red',marker='*')
plt.plot(gas.Year,gas.USA,label='usa',color='green',marker="o")
plt.plot(gas.Year,gas.Italy,label='Italy',color='grey',marker='.')
# if you want to print all the countries you go
#for country in gas:
    #if country!='Year':
        #plt.plot(gas.Year,gas[country],marker='.')
        
## you can add ticks to a graph.here am going to make one of my tick represent 3years.
plt.xticks(gas.Year[::3])
## we can also mark the x and y part of the graph by giving them names
plt.xlabel('Year')
plt.ylabel('us Dollar')

plt.savefig('gas_price_tutorial.png',dpi=300)
plt.legend(gas)
plt.show()
```

```python
## you can mark out the countries you will like to use
intrested=gas.Year
for country in gas:
    if intrested  > 1990:
        plt.plot(gas.Year,gas[country],marker='.')
        plt.plot(gas.Year,gas.Germany,label='Germany',color='red',marker='*')
        plt.plot(gas.Year,gas.USA,label='usa',color='green',marker="o")
        plt.plot(gas.Year,gas.Italy,label='Italy',color='grey',marker='.')
        
        plt.xlabel('Year')
plt.ylabel('us Dollar')


plt.legend()
plt.show()
```

## histogram

```python
bins=(40,50,60,70,80,90,100)

plt.xticks(bins)
plt.ylabel('No of players')
plt.xlabel('skill level')
plt.title('fifa skill rating 2018')
plt.hist(fifa.Overall,bins=bins,color='grey')
plt.show()
```

### pie chart

```python
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
```

```python
right
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
```

```python
left
```

```python
labels=['left','right']
color=['blue','red']

plt.title('foot preference of players in fifa')
plt.pie([left,right],labels=labels,colors=color,autopct=('%.2f %%'))
plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] 
```

```python
fifa.Weight
```

```python
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy=fifa.loc[fifa.Weight>=200].count()[0]
```

```python
weights=[light,light_medium,medium,medium_heavy,heavy]
```

```python
labels=['under 125','125-150','150-175','175-200','200']
explode=[.4,.2,0,0,.4]
plt.pie(weights,labels=labels,autopct='%f %%',pctdistance=0.8,explode=explode)
plt.title('weight of fifa players in(lbs)')
plt.style.use('ggplot')
plt.show()
```

```python
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
CU=fifa.loc[fifa.Club=='Cambridge United']['Overall']
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8),dpi=300)
labels=('FC Barcelona','Real Madrid','CU')
plt.title('fifa football team comparison 2018')
plt.ylabel('fifa overall rating')
plt.savefig('fifa football team comparison 2018.png',dpi=300)
plt.boxplot([barcelona, madrid,CU],labels=labels)
plt.show()
```
