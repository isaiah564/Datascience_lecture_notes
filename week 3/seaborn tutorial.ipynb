# seaborn tutorial

# setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# import data

```python
cs=pd.read_csv('ComputerSales.csv')
print(sns.get_dataset_names())
```

# Distrubution plots

# distrubution plot

```python
sns.distplot(cs['Age'],bins=10)
plt.show()
```

# join plot

```python
## a join plot is used to compare to distrubution
cs
fs=pd.read_csv('icecreamsales.csv')
xc=sns.jointplot(x='Temperature',y='Sales',data=fs,kind='reg')
## you can also make python plot another type of graph
re=sns.jointplot(x='Temperature',y='Sales',data=fs,kind='kde')
s=sns.jointplot(x='Temperature',y='Sales',data=fs,kind='hex')
print(xc)
print(re)
print(s)
```

# kde plot

```python
##Kdeplot is a Kernel Distribution Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables
sns.kdeplot(fs['Temperature'])
```

# pair plots

```python
##Pair Plots are a really simple (one-line-of-code simple!) way to visualize relationships between each variab
vb=sns.pairplot(cs)
vn=sns.pairplot(fs,hue='Sales',palette='Blues')
print(vb)
print(vn)
```

# rug plot

```python
##A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis
sns.rugplot(cs['Profit'])
```

# styling

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,4),dpi=300)
##you can make the fontsize of the x and y label look more visible.note the context must be in paper, notebook, talk, poster
sns.set_context('paper',font_scale=1.5)
sns.jointplot(x='Temperature',y='Sales',data=fs,kind='reg')
```

# categorical plotting

# Bar plot 

```python
## bar plot  shows the relationship between a numeric and a categoric variable.note you can change the way the data can be aggregated by using estimator and the aggregate functions you can use are meann,cov and std
sns.barplot(x='Temperature',y='Sales',data=fs)
```

# count plot

```python
##countplot() method is used to Show the counts of observations in each categorical bin using bars
sns.countplot(x='Sales',data=fs)
```

# box plot

```python
sns.boxplot(x='Sex',y='Age',data=cs,hue='Year')
plt.show()
plt.figure(dpi=300)
```

# violin plot

```python
##A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data.
sns.violinplot(x='Sex',y='Age',data=cs,hue='Year')
```

# strip plot

```python
##A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values.
plt.figure(figsize=(8,5),dpi=300)
sns.stripplot(x='Sex',y='Age',data=cs,jitter=True,hue='Year',dodge=True)
plt.show()
```

# swarm plots

```python
sns.violinplot(x='Sex',y='Age',data=cs)
sns.swarmplot(x='Sex',y='Age',data=cs,color='white')
```

# palette

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')           
sns.stripplot(x='Sex',y='Age',data=cs,hue='Year',palette='gist_heat')
```

# heat maps

```python
##A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours. 
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
fs.mx=fs.corr
sns.heatmap(fs.mx,annot=True,cmap='blues',shape=(0))
```
